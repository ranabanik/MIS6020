#### Task 1: create python (virtual) environment
Python environments are essential for organizing and isolating dependencies, managing packages, and ensuring that your Python code runs consistently across different projects and systems.
To create environment see: https://tinyurl.com/58e38cxw

For example, I created Python environment called `reg37` that runs Python code version 3.7. <br>
**(base) banikr@banikr:~\$** `conda create --name reg37 python=3.7` <br>
To check if the environment there or other environments:  <br>
**(base) banikr@banikr:~\$** `conda info --envs`
 

Now activate the `reg37` environment: <br>
**(base) banikr@banikr:~\$** `conda activate reg37` <br>
Once the `reg37` environment is activated, `base` will change into `reg37`, which means you are now in `reg37` environment. 


#### Task 2: install packages
Now we will install some amazing Python libraries/packages which we will use for registration. But before that check the built-in packages by running `pip list` command: <br>
**(reg37) banikr@banikr:~\$** `pip list` <br>
`pip` is a package manager for Python. 
Now install the following packages one by one (after each line the dependencies will be downloaded and installed...): <br>
**(reg37) banikr@banikr:~\$** `pip install SimpleITK` <br>
**(reg37) banikr@banikr:~\$** `pip install nibabel` <br>
**(reg37) banikr@banikr:~\$** `pip install numpy` <br>
**(reg37) banikr@banikr:~\$** `pip install matplotlib` <br>
**(reg37) banikr@banikr:~\$** `python -m pip install -U scikit-image` <br>
**(reg37) banikr@banikr:~\$** `pip install jupyter` <br>

Now run `pip list` again. You shall see the packages above and their versions. Take a screenshot.
